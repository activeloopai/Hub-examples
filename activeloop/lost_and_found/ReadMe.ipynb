## Imports

```python
import matplotlib.pyplot as plt
import hub
```

## Load the dataset

```python
ds = hub.load("activeloop/lost_and_found_semantic_segmentation_train")
ds_test = hub.load("activeloop/lost_and_found_semantic_segmentation_test")
```

## Visualize

```python
def visualize(image, mask):
    plt.figure(figsize=(8, 6), dpi=80)
    plt.subplot()
    plt.imshow(image)
    plt.imshow(mask, alpha=0.3)

    plt.show()
```

```python
visualize(ds['image_left', 5].compute(), ds['segmentation_label', 5].compute())
visualize(ds_test['image_left', 5].compute(), ds_test['segmentation_label', 5].compute())
```
